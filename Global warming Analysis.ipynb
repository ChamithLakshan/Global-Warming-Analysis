# Is there Global Warming

```python
global_warm=pd.read_csv('GlobalTemperatures.csv')
global_warm.head()
```

```python
global_warm['dt'][0].split('-')[0]#acsess the year
```

```python
def fetch_year(date):
    return date.split('-')[0]
```

```python
global_warm['years'] = global_warm['dt'].apply(fetch_year)
```

```python
global_warm.head()
```

```python
data=global_warm.groupby('years').agg({'LandAverageTemperature':'mean','LandAverageTemperatureUncertainty':'mean'}).reset_index()
data.head()
```

```python
data['Uncertanity Top']=data['LandAverageTemperature'] + data['LandAverageTemperatureUncertainty']
data['Uncertanity Bottom']=data['LandAverageTemperature'] - data['LandAverageTemperatureUncertainty']
```

```python
data.head()
```

```python
data.columns
```

```python
fig=px.line(data,x='years' ,y=['LandAverageTemperature','Uncertanity Top', 'Uncertanity Bottom'], title='Average Land Temperature in World')
fig.show()
```

<h1>Average temperature in Each season</h1>

```python
global_warm.head()
```

```python
global_warm['dt'].dtype
```

```python
global_warm['dt'] = pd.to_datetime(global_warm['dt'])
```

```python
global_warm['month'] = global_warm['dt'].dt.month
```

```python
global_warm.head()
```

```python
def get_season(month):
    if month>=3 and month<=5:
        return 'spring'
    elif month>=6 and month<=8:
        return 'summer'
    elif month>=9 and month<=11:
        return 'autumn'
    else:
        return 'winter'
```

```python
global_warm['season'] = global_warm['month'].apply(get_season)
```

```python
global_warm.head()
```

```python
years = global_warm['years'].unique()
```

```python
spring_temps=[]
summer_temps=[]
autumn_temps=[]
winter_temps=[]
```

```python
for year in years:
    current_df=global_warm[global_warm['years']==year]
    spring_temps.append(current_df[current_df['season']=='spring']['LandAverageTemperature'].mean())
    summer_temps.append(current_df[current_df['season']=='summer']['LandAverageTemperature'].mean())
    autumn_temps.append(current_df[current_df['season']=='autumn']['LandAverageTemperature'].mean())
    winter_temps.append(current_df[current_df['season']=='winter']['LandAverageTemperature'].mean())
```

```python
spring_temps
```

```python
season =pd.DataFrame()
```

```python
season['year']=years
season['spring_temps']=spring_temps
season['summer_temps']=summer_temps
season['autumn_temps']=autumn_temps
season['winter_temps']=winter_temps
```

```python
season.head()
```

```python
season.columns
```

```python
fig=px.line(season,x='year',y=['spring_temps', 'summer_temps', 'autumn_temps', 'winter_temps'],title='Avg temperature in each season')
fig.show()
```

# Average Temperature of Sri lankan cities

```python
cities = pd.read_csv('GlobalLandTemperaturesByCity.csv')
cities.head()
```

```python
SL=cities[cities['Country']=='Sri Lanka']
```

```python
SL.shape
```

```python
SL['City'].unique()
```

```python
SL.head()
```

```python
SL['Latitude']=SL['Latitude'].str.strip('N')
SL['Longitude']=SL['Longitude'].str.strip('E')
```

```python
SL.head()
```

```python
SL['dt'] = pd.to_datetime(SL['dt'])
```

```python
SL['month']= SL['dt'].dt.month
```

```python
SL.head()
```

```python
Sl_temp=SL.groupby(['month','City'])['AverageTemperature'].mean().to_frame().reset_index()
```

```python
Sl_temp.head()
```

```python
df=Sl_temp.merge(SL,on='City')
```

```python
df.head()
```

```python
data=df.drop_duplicates(subset=['month_x','City'])
data.head()
```

```python
import plotly.graph_objs as go
```

```python
data1=go.Heatmap(x=data['month_x'],y=data['City'],z=data['AverageTemperature_x'])
```

```python
layout=go.Layout(title='Avg temp of sl cities by month')
```

```python
fig=go.Figure(data=data1,layout=layout)
fig.show()
```
